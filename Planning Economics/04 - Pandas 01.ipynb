## Read data

### - Human development index (HDI)
<b> Source:  [UNITED NATIONS DEVELOPMENT PROGRAMME (Human Development Reports)](http://hdr.undp.org/en/indicators/137506#)

```python
df = pd.read_excel('Human development index (HDI).xlsx', sheet_name='HDI_1')
df
```

```python
df.columns
```

```python
df.index
```

```python
df.shape
```

```python
df.head()
```

```python
df.head(10)
```

```python
df = df.rename(columns={'Iran (Islamic Republic of)':'Iran', 'Russian Federation':'Russian' })
df
```

```python
df.plot(x='year', y = 'Iran',
       kind = 'line',
        figsize = (15,7), fontsize = 16,
        color = 'red',
        linewidth = 3
       );

plt.grid(ls = '--', c= 'gray')
plt.xlabel('Year', fontsize = 14, fontname = 'Times New Roman')
plt.ylabel('HDI', fontsize = 14, fontname = 'Times New Roman')
plt.title("Iran's Human Develpment Index Since\n1990 t0 2019",fontsize = 16, fontname = 'Times New Roman')
plt.legend(loc = 'best', fontsize = 13)
plt.xlim(left = 1990, right = 2020)
plt.show()
```

```python
df.columns
```

```python
df.plot(x='year', y = ['Iran','Australia', 'Canada', 'China','United States', 'Russian'],
       kind = 'line',
        figsize = (15,7), fontsize = 16,
        #color = 'red',
        linewidth = 3
       );

plt.grid(ls = '--', c= 'gray')
plt.xlabel('Year', fontsize = 14, fontname = 'Times New Roman')
plt.ylabel('HDI', fontsize = 14, fontname = 'Times New Roman')
plt.title("Human Develpment Index Since\n1990 t0 2019",fontsize = 16, fontname = 'Times New Roman')
plt.legend(loc = 'best', fontsize = 13)
plt.xlim(left = 1990, right = 2020)
#plt.savefig('HDI_1.jpg', dpi = 700)
plt.show()
```
